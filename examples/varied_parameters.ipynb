## Varied Parameters
This notebook contains code to help visualize how varying parameters of the DK_14 profiles can change them.
Place this Sam's splashback_tools/examples so fits_profs imports correctly, and use the multinest output for sample_thetas

```python
from splashback_tools.profiles import fit_d22
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_thetas = np.load("fit_Mbin1.npy")
```

```python
#Example of parameters
base_theta = np.array([-0.50766891,  0.6385919 ,  2.82672365, -0.62170237, -0.17964018,
        0.99961697,  0.17812577,  2.34861028, -3.10378803])
```

```python

def plot(param_index, theta, sample_thetas):
    r_thr = np.logspace(-1, np.log10(7), 100)
    theta = np.copy(theta)
    dim = len(sample_thetas[0])
    plt.xscale('log')
    plt.yscale('log')
    for i, param in enumerate(np.sort(sample_thetas[param_index])):
        theta[param_index] = param
        percentile = i/dim
        plt.plot(r_thr, fit_d22.rho_D22(theta, r_thr), color = (percentile, 1 - percentile, 1), linewidth = 0.7)
```

```python
#Alpha plot
plot(0, base_theta, sample_thetas)
```

```python
#Beta plot
plot(1, base_theta, sample_thetas)
```

```python
#Gamma plot
plot(2, base_theta, sample_thetas)
```

```python
#r_s plot
plot(3, base_theta, sample_thetas)
```

```python
#r_t plot
plot(4, base_theta, sample_thetas)
```

```python
#d_1 plot
plot(5, base_theta, sample_thetas)
```

```python
#s plot
plot(6, base_theta, sample_thetas)
```

```python
#d_max plot
plot(7, base_theta, sample_thetas)
```
